## Waterfall chart function

The fuction is from https://github.com/chrispaulca/waterfall
Starts with $ pip install waterfallcharts

```python
%matplotlib inline
import waterfall_chart

a = ['sales','returns','credit fees','rebates','late charges','shipping']
b = [10,-30,-7.5,-25,95,-7]

my_plot = waterfall_chart.plot(a, b)
```

```python
my_plot = waterfall_chart.plot(a,b,rotation_value=90, Title="Waterfall chart", green_color='#000000', x_lab="Time", y_lab="Amount in '000",net_label="End Result")
```
